#### Slope

$ change  in  y / change  in  x $

how do we measure the rate of change x has on y? 

> we measure the ***tangent line*** at x. this gives us the ***instantaneous slope***, which is the derivative

we can approximate the derivative of a function at x by using this point and another one also taken at x, but with a very small delta added to it, such as 0.0001.

> you do not want the delta to be too big

> this lets us perform the same calculation for the slop as before, but on two points that are very close to each other, resulting in the approximation of the slope at x:

```python
p2_delta = 0.00001

x1 = 1
x2 = x1 + p2_delta

def f(x):
    return 2*x**2

y1 = f(x1) #the result at the derivation point
y2 = f(x2) #the reuslt at the other derivation point

approx_derivative = (y2-y1) / (x2-x1)
print(approx_derivative)
```

#### The Numerical Derivative

the method of calculating the derivative is the ***numerical differentation***, which is calculating the slope of a tangent line using two infinitely close points

> computing the slope of a tangent line made from two points that were close.

> the closer the two points are to each other, the more correct the tangent line will be.

lets visualize the tangent lines and how they change depending on where we calculate them

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x**2

#np.arange(start, stop, step) to give use a smoother line
x = np.arange(0, 5, 0.0001)
y = f(x)

plt.plot(x, y)
plt.show()
```

so, to draw these tangent lines, we will derive the function for the tangent line at a point, and plot the tangent on the graph:

function for a straight line: $ y = mx + b $ 

to calculate $b$, the formula is $ b = y - mx $

now, given the above equation for $b$, the approximation of the derivative and the same "close enough" point (its x and y coordinates), we can substitute them in the equation and get the y-intercept for the tangent line at the derivation point

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x**2

#np.arange(start, stop, step), to give us a smaller line
x = np.arange(0,5,0.001)
y = f(x)

plt.plot(x, y)
plt.show()

#the point and the close enough point
p2_delta = 0.0001
x1 = 2
x2 = x1+p2_delta

y1 = f(x1)
y2 = f(x2)

print((x1, y1), (x2, y2))

#derivative approximation and the y-intercept for the tangent line
approx_derivative = (y2-y1) / (x2-x1)
print(approx_derivative)
b = y2 - approx_derivative * x2
print(b)

'''we put the tangent line calculation into a function so that we can call it 
    multiple times for the different values of x
    
    approximate derivative and b are constant for the given function, thus 
    calculated once above this function.'''
def tangent_line(x):
    return approx_derivative*x + b   

'''plotting the tangent line
    +/- 0.9 to draw the tangent line on our graph
    then we calculate teh y for given x using the tangent line function '''
to_plot = [x1-0.9, x1, x1+0.9] 
plt.plot(to_plot, [tangent_line(i) for i in to_plot])

print('Approximate derivative for f(x): ', f'where x = {x1} is {approx_derivative}')

plt.show()
```

we care about the slope of the tangent line at that x because it tells us the impact that x has on this function at this particular point.

We will use this concept to determine the effect of a specific weight or bias on the overall loss function given a sample.
